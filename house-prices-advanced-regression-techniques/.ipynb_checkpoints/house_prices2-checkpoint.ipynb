# Ridge

```python
from sklearn.linear_model import Ridge, Lasso, ElasticNet
from sklearn.model_selection import GridSearchCV
```

```python
ridge = Ridge()

ridge.set_params(max_iter=100, normalize=True)
param_ridge_grid = {'alpha': np.linspace(1e-3, 10, 500)}
ridge_gridsearch = GridSearchCV(ridge, param_grid=param_ridge_grid, cv=10, return_train_score=True, n_jobs=-1)

ridge_gridsearch.fit(X_train, y_train)
```

```python
ridge_gridsearch.best_params_ 
```

```python
ridge_gridsearch.best_score_  #best cv score #score=R2   -0.2  #cross-validation score
```

```python
ridge_gridsearch.score(X_test, y_test)  #cv score -0.1 (better)
```

```python
ridge_gridsearch_rmse = GridSearchCV(ridge, param_grid=param_ridge_grid, cv=10, return_train_score=True, scoring='neg_root_mean_squared_error', n_jobs=-1)

ridge_gridsearch_rmse.fit(X_train, y_train)  #finding root mean square error 
```

```python
ridge_gridsearch_rmse.best_params_ 
```

```python
ridge_gridsearch_rmse.best_score_ 
```

```python
ridge_gridsearch_rmse.score(X_test, y_test)
```

```python
#refit the model with the best hyperparameter

ridge2 = Ridge(alpha=0.22141883767535073, normalize=True)
ridge2.fit(X_train,y_train)
pred = ridge2.predict(X_test)

from sklearn.metrics import mean_squared_error
mean = mean_squared_error(y_test, pred) 
rmse = np.sqrt(mean_squared_error(y_test,pred))

rmse
```
